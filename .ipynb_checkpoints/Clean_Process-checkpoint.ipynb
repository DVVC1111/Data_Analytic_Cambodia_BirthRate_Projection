```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('cambodia-population-2023-11-20.csv')

# Filter the data for the desired range (2000 to 2020)
df_filtered = df[(df['date'] >= '2000-01-01') & (df['date'] <= '2020-12-31')]

# Display the filtered DataFrame
print(df_filtered)
```

```python
import pandas as pd

# Assuming your dataset is stored in a CSV file named "IDB5YR.IDB5YEAR-2023-11-21T140834.csv"
data = pd.read_csv("IDB5YR.IDB5YEAR-2023-11-21T140834.csv")

start_year = 2000
end_year = 2023

# Filter the data between 2000 and 2023
filtered_data = data[(data['Year (time)'] >= start_year) & (data['Year (time)'] <= end_year)]

# Select the specific factors of interest including 'Year (time)'
factors_of_interest = ['Year (time)',
                       'Crude birth rate (births per 1,000 population) (CBR)',
                       'Total fertility rate (TFR)',
                       'Net migration rate (net number of migrants per 1,000 population) (NMR)',
                       'Population density (total population divided by area in square kilometers) (POP_DENS)'
                       ]

# Filter the dataset to include only the selected factors
filtered_data = filtered_data[factors_of_interest]

# Convert the filtered data to a dataframe
filtered_dataframe = pd.DataFrame(filtered_data)

# Print the filtered DataFrame
print(filtered_dataframe)
```
